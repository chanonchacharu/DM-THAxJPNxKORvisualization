```python
import pandas as pd
import numpy as np
import json
import os

drama_db = {}

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        path = os.path.join(dirname, filename)
        df = pd.read_csv(path)
        
        key = str(filename).split('.')[0].strip()
        drama_db[key] = df
        
        print(f'{key=}')
```
